Changing Names

```python
df = df.rename(columns={'Do you believe that a hotdog falls under the category of a sandwich?': 'Sandwich?'})
df = df.rename(columns={'What is the rationale behind your opinion?': 'Reason'})
df = df.rename(columns={'Are you currently enrolled as an undergraduate or graduate student at the University of Virginia?': 'Student Status'})
df = df.rename(columns={'What is your field of study or major?': 'Major'})
df = df.rename(columns={'What state or country were you born in?': 'State Born'})
df = df.rename(columns={'If you are an undergraduate, what is your current academic year?': 'Year at School'})
df = df.rename(columns={'Gender: How do you identify?': 'Gender'})
df = df.rename(columns={'You preferred to self-describe for gender; please, share your identity below': 'Gender Description'})
```

Cleaning the "sandwich" variable

```python
df['Sandwich?'] = df['Sandwich?'].str[0]
df['Sandwich?'] = df['Sandwich?'].str.upper()

df.value_counts("Sandwich?")
```

```python
df.head()
```

```python
import seaborn as sns

# Column names
df.columns = ['Timestamp', 'Response', 'Rationale', 'Enrolled', 'Major', 'State', 'Year', 'Gender', 'GenderDescribe']

# Response column
df['Response'] = df['Response'].apply(lambda x: 'Yes' if x[0].lower() == 'y' else x)
df['Response'] = df['Response'].apply(lambda x: 'No' if x[0].lower() == 'n' else x)

# State column
df['State'] = df['State'].apply(lambda x: 'Virginia' if 'virginia' in str(x).lower() else x)
df['State'] = df['State'].apply(lambda x: 'Virginia' if 'va' in str(x).lower() else x)
df['State'] = df['State'].apply(lambda x: 'New Jersey' if 'nj' in str(x).lower() else x)
df['State'] = df['State'].apply(lambda x: 'Washington D.C.' if 'dc' in str(x).lower() else x)
df['State'] = df['State'].apply(lambda x: 'New York' if 'ny' in str(x).lower() else x)
df['State'] = df['State'].apply(lambda x: 'California' if 'ca' in str(x).lower() else x)
df['State'] = df['State'].str.title()
df['State'] = df['State'].apply(lambda x: 'United States' if 'us' in str(x).lower() else x)
df['State'] = df['State'].apply(lambda x: 'United States' if 'america' in str(x).lower() else x)
df['State'] = df['State'].apply(lambda x: 'United States' if 'u.s.' in str(x).lower() else x)

# Removing rows and columns
df = df.drop(['Timestamp', 'Rationale', 'GenderDescribe'], axis=1)
df = df.dropna()
df = df.drop('Enrolled', axis=1)

# Gender column
df['Gender'] = df['Gender'].apply(lambda x: 'Other' if 'prefer' in str(x).lower() else x)
df['Gender'] = df['Gender'].apply(lambda x: 'Other' if 'non-binary' in str(x).lower() else x)

# Major column
df['Major'] = df['Major'].str.title()

display(df)


```

```python
counts = df.groupby(['Gender', 'Response']).size().reset_index(name='Count')
counts
```

```python
import matplotlib.pyplot as plt

sns.barplot(data=counts, x='Response', y='Count', hue='Gender', palette='viridis')

plt.title('Yes and No Responses Split by Gender')
plt.xlabel('Sandwich')
plt.ylabel('Count')
```

```python
counts = df.groupby(['Year', 'Response']).size().reset_index(name='Count')


category_order = ["First year", "Second year", "Third year","Fourth year"]
counts["Year"] = pd.Categorical(counts["Year"], categories=category_order, ordered=True)

counts
```

```python
sns.barplot(data=counts, x='Year', y='Count', hue='Response', palette='viridis')

plt.title('Yes and No Responses Split by Year in School')
plt.xlabel('Sandwich')
plt.ylabel('Count')
```
